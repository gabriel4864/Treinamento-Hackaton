```python
import pandas as pd
import numpy as np

# Carregar os arquivos Excel
df1 = pd.read_excel("Despesas.xlsx")
df2 = pd.read_excel("Orçamento.xlsx")
```

```python
print(df1.dtypes)
```

```python
# Convertendo a coluna de milissegundos para data
df1['Data'] = pd.to_datetime(df1['Data'], unit='ms')
df1.drop(columns='data', inplace=True)

# Exibindo o DataFrame resultante
display(df1)
```

```python
meses = {
    1: 'janeiro', 2: 'fevereiro', 3: 'março', 4: 'abril',
    5: 'maio', 6: 'junho', 7: 'julho', 8: 'agosto',
    9: 'setembro', 10: 'outubro', 11: 'novembro', 12: 'dezembro'
}

df2['Mes'] = df2['Mes'].map(meses)

display(df2)
```

```python
# Criar o escritor do Excel
with pd.ExcelWriter("Financiamento.xlsx", engine='xlsxwriter') as writer:
    df1.to_excel(writer, sheet_name="Planilha1", index=False)
    df2.to_excel(writer, sheet_name="Planilha2", index=False)
```
